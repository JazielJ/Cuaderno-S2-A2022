


````python
### Codigo
"""
Nombre: ejercicio_13
Entrada: un numero
Salida: su factorial
Restricciones: Debe ser entero y no negativo
"""
def factorial(num):
    if isinstance(num, int):
        if(num < 0):
            return "Error debe ser mayor cero"

    if (num <= 0):
        return 1
    fac = 1
    while (num > 0):
        fac = fac * num
        num = num - 1
    return fac
````
