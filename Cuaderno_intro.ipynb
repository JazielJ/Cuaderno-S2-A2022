### totalDigitosDivisibles
Desarrolle un algoritmo para determinar cuantos digitos de un numero son divisibles entre un divisor designado

````python
### Codigo python
"""
Nombre:totalDigitosDivisibles
Entrada:Un numero y un dvisor
Salida:Los digitos divisibles entre el divisor
Restricciones:Ambos parametros deben ser enteros y mayores a cero
"""
def totalDigitosDivisibles(num, divisor):
    exponente = 0
    resultado = 0
    if isinstance(num, int) and isinstance(divisor, int):
        if(divisor > 0) and (divisor < 10):
            while(num > 0):
                digito = num % 10
                if(digito % divisor == 0):
                    resultado += digito * 10**exponente
                    exponente += 1
        
                num = num // 10
            return resultado
        else:
            return "Error: divisor debe estar entre 0 y 10"
    else:
        return "Error: Num y divisor deben ser tipo entero"
````
