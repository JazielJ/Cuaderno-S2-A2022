````python
### Codigo
"""
Nombre: ejercicio_7
Entrada: un precio
Salida: monto a pagar
Restricciones: todos los trajes que
tienen un precio superior a ¢2500 se les
aplicará un descuento de 15 %, a todos
los demás se les aplicará sólo 8 %.
"""
def ejercicio_7(p):
    des = 0.08
    Des = 0.15
    total = p * des
    Total = p * Des
    if(p < 2500 ):
        Pago = p - total
        print(Pago)
    if(p >= 2500):
        Pago = p - Total
        print(Pago)
````

### Ejercicio 9
Se requiere un algoritmo dado un número si este es par o impar.
Como restricción el número debe ser entero y mayor o igual a cero

### Diagrama
Ejercicio 9.png

### Pseudocodigo


```

	Leer num

		Escribir True
	SiNo
		Escribir False
	FinSi
FinAlgoritmo
```



````python
### Codigo
"""
Nombre: ejercicio_9
Entrada: un numero
Salida: true o false
Restricciones:  el valor debe ser Entero y mayor a CERO
"""
def ejercicio_9(num):
    if isinstance(num, int):
        if(num % 2 == 0):
            print("es par")
        if(num % 2 != 0):
            print("no es par")
````
